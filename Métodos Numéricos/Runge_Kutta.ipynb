```python
def f1(x, y):
  return x+y
```

```python
h = 0.2
xo = 0
yo = 1
```

```python
import numpy as np
X = np.linspace(start = xo, stop = 10, num=100)
X
```

```python
import numpy as np
def RK4(xo, yo, h):
  global X, Y
  X = [xo]
  Y = [yo]

  for i in range(0,100):
    X.append(X[i]+h)
    p = f1(X[i], Y[i])*h
    q = f1(X[i]+0.5*h, Y[i]+0.5*p)*h
    r = f1(X[i]+0.5*h, Y[i]+0.5*q)*h
    s = f1(X[i]+h, Y[i]+r)*h
    Y.append(Y[i]+(1/6)*(p+2*q+2*r+s))
  return X, Y
```

```python
RK4(xo= 0, yo = 1, h = 0.2)
```

```python
X
```

```python
X = np.array(X)
Yreal = -1 -X +2*np.exp(X)

import pandas as pd
df = pd.DataFrame(list(zip(X, Y, Yreal)), columns =['X', 'Y', 'Yreal'])
df
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y, 'o', label = 'RK4')
plt.plot(X, Yreal, label ='Soluci√≥n Real')
plt.grid()
plt.legend()
plt.title('RK4 vs Soluci√≥n Real')
plt.xlabel('X')
plt.ylabel('Funci√≥n Soluci√≥n Y')
plt.show()
```

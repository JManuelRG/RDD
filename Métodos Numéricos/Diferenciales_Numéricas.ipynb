```python
import math

def f(x):
  return math.exp(-2*x)-x
```

```python
# Primera derivada centrada con dos puntos
(f(2.5)-f(1.5))/(2.5-1.5)
```

```python
# Evaluación de la primera derivada:
def der(x, h):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
h = [0.5, 0.1, 0.001, 0.0001, 0.00001, 1e-6]
for i in range(len(h)):
  print(f"Paso = {h[i]}, f'(2)={der(x=2, h=h[i])}")
```

```python
import matplotlib.pyplot as plt
plt.plot(h, [der(x=2, h=h[i]) for i in range(len(h))])
plt.title("f'(2) aplicando la diferencia finita centrada")
plt.xlabel("Tamaño de paso, h")
plt.ylabel("Valor de la primera derivada")
plt.axhline(y = -1-2/math.exp(4), color = 'r', linestyle = '--')
plt.grid()
plt.show()
```

```python
import math

def f(x):
  return math.exp(-2*x)-x
```

```python
# Diferencias finitas hacia adelante
def dfad(x, h):
  return (f(x+h)-f(x))/(h)
```

```python
dfad(x=2, h=0.5) # hacia adelante
```

```python
# Diferencia de primer orden, hacia adelante
h = [0.5, 0.1, 0.001, 0.0001, 0.00001, 0.000001]

for i in range(len(h)):
  print(f"Paso = {h[i]}, f'(2)= {dfad(x=2, h=h[i])}")
```

```python
import matplotlib.pyplot as plt
plt.plot(h, [dfad(x=2, h=h[i]) for i in range(len(h))])
plt.title("f'(2) aplicando la diferencia finita hacia adelante, orden 1")
plt.xlabel("Tamaño de paso, h")
plt.ylabel("Valor de la primera derivada")
plt.axhline(y = -1-2/math.exp(4), color = 'r', linestyle = '--')
plt.grid()
plt.show()
```

```python
# Diferencia de segundo orden, hacia adelante
def dfad2(x, h):
  return (-f(2+2*h)+4*f(2+h)-3*f(2))/(2*h)
```

```python
dfad2(x=2, h=0.5)
```

```python
# Diferencia de segundo orden, hacia adelante
h = [0.5, 0.1, 0.001, 0.0001, 0.00001, 0.000001]

for i in range(len(h)):
  print(f"Paso = {h[i]}, f'(2)= {dfad2(x=2, h=h[i])}")
```

```python
import matplotlib.pyplot as plt
plt.plot(h, [dfad2(x=2, h=h[i]) for i in range(len(h))])
plt.title("f'(2) aplicando la diferencia finita hacia adelante, orden 2")
plt.xlabel("Tamaño de paso, h")
plt.ylabel("Valor de la primera derivada")
plt.axhline(y = -1-2/math.exp(4), color = 'r', linestyle = '--')
plt.grid()
plt.show()
```

# Diferencia Finita hacia atrás 1er orden

# Diferencia Finita hacia atrás, segundo orden

```python
def dfat2(x, h):
  return (3*f(x)-4*f(x-h)+f(x-2*h))/(2*h)
```

```python
# Diferencia de segundo orden, hacia atrás
h = [0.5, 0.1, 0.001, 0.0001, 0.00001, 0.000001]

for i in range(len(h)):
  print(f"Paso = {h[i]}, f'(2)= {dfat2(x=2, h=h[i])}")
```

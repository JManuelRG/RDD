```python
def Lagrange(X):
  f3 = yo*(X-x1)*(X-x2)*(X-x3)/((xo-x1)*(xo-x2)*(xo-x3)) + y1*(X-xo)*(X-x2)*(X-x3)/((x1-xo)*(x1-x2)*(x1-x3)) + y2*(X-xo)*(X-x1)*(X-x3)/((x2-xo)*(x2-x1)*(x2-x3)) + y3*(X-xo)*(X-x1)*(X-x2)/((x3-xo)*(x3-x1)*(x3-x2))
  return f3
```
